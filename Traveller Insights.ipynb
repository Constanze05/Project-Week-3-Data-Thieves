```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = 'https://www.cntraveller.com/gallery/best-holiday-destinations-2020'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")  
soup
```

```python
soup2=soup.find_all('div', attrs={'class':'c-figure__title'})
soup2
```

```python
ranking = [element.text for element in soup2]
ranking
```

```python
ranking_df = pd.DataFrame(ranking)
ranking_df
```
