```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_busiest_airports_by_passenger_traffic#2018_statistics'
html = requests.get(url).content
#html

soup = BeautifulSoup(html, "lxml")
#soup.find('tr')

# Search for tables in the webpage
table = soup.find_all('table', attrs ={'style':'font-size: 85%'})
#table

# There are multiple tables, the second one (table[1]) is the one for 2018
rows = table[1].find_all('tr')
#rows
```

```python
# Cleaning the data 
rowsss = [row.text.replace('\n', '!') for row in rows[1:]]
#rowsss


result = [x.split('!') for x in rowsss]
#result
```

```python
# Separate different data into different columns 
airports =  [result[i][2] for i in range(len(result))]
#airports

cities =  [result[i][3] for i in range(len(result))]
#cities

countries =  [result[i][4] for i in range(len(result))]
#countries

codes =  [result[i][5] for i in range(len(result))]
#codes

code = [code.split("/") for code in codes]
icao = [code[i][1] for i in range(len(code))]
city_code = [code[i][0] for i in range(len(code))]

passengers =  [result[i][6] for i in range(len(result))]
passengers = [int(value.replace(",","")) for value in passengers]
#passengers
```

```python
# Combine the data into one DataFrame
ds=pd.DataFrame()

ds["Airport"]=airports
ds["City"]=cities
ds["Country"]=countries
ds["City Code"]=city_code
ds["ICAO"]=icao
ds["Passengers"]=passengers
```
